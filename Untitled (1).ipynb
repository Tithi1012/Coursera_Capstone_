```python
import numpy as np
arr= np.full([181, 3], None)
```

```python
#MAKING AN ARRAY
n1=0
for r1 in r:
    r2= r1.find_all('td')
    n2=0
    for r3 in r2:
        arr[n1][n2]=str(r3)
        n2=n2+1
    n1=n1+1
```

```python
def category_decide (row):
    try:
        category = row['categories']
    except:
        category = row['venue.categories']
        
    if len(category) == 0:
        return None
    else:
        return category[0]['name']
```

* *FINALLY, WE CAN NOW GET THE COLUMNS* *

```python
venues = results['response']['groups'][0]['items']
    
nearby = pd.json_normalize(venues) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby =nearby.loc[:, filtered_columns]

# filter the category for each row
nearby['venue.categories'] = nearby.apply(category_decide, axis=1)

# clean columns
nearby.columns = [col.split(".")[-1] for col in nearby.columns]

nearby.head()
```
